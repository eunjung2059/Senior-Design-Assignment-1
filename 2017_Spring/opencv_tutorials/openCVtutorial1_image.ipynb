###  [Image](http://docs.opencv.org/3.1.0/dc/d2e/tutorial_py_image_display.html)
how to read an image, how to display it and how to save it back

    cv2.imread()
    cv2.imshow()
    cv2.imwrite()

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

    cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.
    cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode
    cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel

1, 0 , -1 can be used for each flag

```python
# read an image
img = cv2.imread('hand.png', cv2.IMREAD_GRAYSCALE) # apply gray filter
```

```python
# display an image
cv2.imshow('image', img)
```

```python
cv2.waitKey(0)
```

```python
# close all windows
cv2.destroyAllWindows()
```

```python
# display using matplotlib.pyplot
plt.imshow(img, cmap='gray', interpolation='bicubic')
```

```python
plt.plot([50, 100], [80, 100], 'c', linewidth=5)
```

```python
plt.show()
```

```python
# save the gray filtered image
cv2.imwrite('grayhand.png', img)
```
